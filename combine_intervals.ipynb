```python
import pandas as pd
import os
```

```python
ROOT_DIR  = 'data/intervals/'

```

```python
intervals = os.listdir(ROOT_DIR)
dfs = []
pr_medians = []
pj_medians = []
p_durations = []
for i in intervals:
    file_path = os.path.join(ROOT_DIR, i)
    df = pd.read_csv(file_path)[['pr_interval', 'pj_interval', 'duration']]
    pr_medians.append(df['pr_interval'].median())
    pj_medians.append(df['pj_interval'].median())
    dfs.append(df)
```

```python
all_intervals = pd.concat(dfs)
```

```python
all_intervals.to_csv("all_intervals.csv")
```
